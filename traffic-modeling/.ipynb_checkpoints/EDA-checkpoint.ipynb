# PeMS-BAY

```python
distances = pd.read_csv('data/third_party/PeMS-BAY/distances_bay_2017.csv', header=None, names=['from', 'to', 'distance'])
distances
```

```python
locations = pd.read_csv('data/third_party/PeMS-BAY/graph_sensor_locations_bay.csv', header=None, names=['sensor_id', 'x', 'y'])
locations
```

```python
plt.scatter(locations['x'], locations['y'], marker='.')
```

```python
sensor_ids, sensor2node, adjastency_matrix = pd.read_pickle('data/third_party/PeMS-BAY/adj_mx_bay.pkl')
```

```python
plt.figure(figsize=(16, 14))
plt.matshow(adjastency_matrix, fignum=1)
plt.colorbar()
```

```python
locations['node_id'] = locations['sensor_id'].astype(str).map(sensor2node)
locations.sort_values('node_id', inplace=True)
locations
```

```python
network = nx.from_numpy_matrix(adjastency_matrix)
```

```python
plt.figure(figsize=(10, 10))
nx.draw(network, pos=locations[['x', 'y']].values, node_size=2)
```

Remove self-connections:

```python
for u in network.nodes():
    network.remove_edge(u, u)
```

Remove all 3-node cycles:

```python
for cycle in nx.cycle_basis(network):
    if len(cycle) <= 3:
        cycle = network.subgraph(cycle)
        _, u, v = min((cycle[u][v]['weight'], u, v) for (u, v) in cycle.edges())
        network.remove_edge(u, v)
```

```python
plt.figure(figsize=(10, 10))
nx.draw(network, pos=locations[['x', 'y']].values, node_size=2)
```

```python
for cycle in nx.cycle_basis(network):
    if len(cycle) <= 4:
        cycle = network.subgraph(cycle)
        _, u, v = min((cycle[u][v]['weight'], u, v) for (u, v) in cycle.edges())
        network.remove_edge(u, v)
```

```python
plt.figure(figsize=(10, 10))
nx.draw(network, pos=locations[['x', 'y']].values, node_size=2)
```

```python
for cycle in nx.cycle_basis(network):
    if len(cycle) <= 35:
        cycle = network.subgraph(cycle)
        _, u, v = min((cycle[u][v]['weight'], u, v) for (u, v) in cycle.edges())
        network.remove_edge(u, v)
```

```python
plt.figure(figsize=(10, 10))
nx.draw(network, pos=locations[['x', 'y']].values, node_size=2)
```
