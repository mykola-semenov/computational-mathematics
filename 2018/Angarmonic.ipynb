# Маятник, метод Эйлера

```python
import numpy, matplotlib.pyplot
```

```python
m = 1.0
g = 9.8
l = 1.0
J = 1.0
```

```python
w = m*g*l/J
```

```python
T = 10
dt = 1e-2
n = int(T/dt)
```

```python
phi = numpy.zeros(n)
V = numpy.zeros(n)
```

```python
phi[0] = numpy.pi/4
V[0] = 4
```

Первый вариант:
$$\phi_{n+1} = \phi_n + \phi'_n\cdot dt$$
$$\phi'_{n+1} = \phi'_n - \omega^2\cdot\sin{\phi_{n+1}}\cdot dt$$

```python
for i in range(1, n):
    phi[i] = phi[i - 1] + V[i - 1]*dt
    V[i] = V[i - 1] - w*numpy.sin(phi[i])*dt
```

```python
matplotlib.pyplot.figure(figsize=(15, 10))
matplotlib.pyplot.plot(phi, V, 'k')
matplotlib.pyplot.grid(True)
matplotlib.pyplot.axis('tight')
matplotlib.pyplot.xlabel('phi')
matplotlib.pyplot.ylabel('V')

matplotlib.pyplot.show()
```

Второй вариант:
$$\phi_{n+1} = \phi_n + \phi'_n\cdot dt$$
$$\phi'_{n+1} = \phi'_n - \omega^2\cdot\sin{\phi_{n}}\cdot dt$$

```python
for i in range(1, n):
    phi[i] = phi[i - 1] + V[i - 1]*dt
    V[i] = V[i - 1] - w*numpy.sin(phi[i - 1])*dt
```

```python
matplotlib.pyplot.figure(figsize=(15, 10))
matplotlib.pyplot.plot(phi, V, 'k')
matplotlib.pyplot.grid(True)
matplotlib.pyplot.axis('tight')
matplotlib.pyplot.xlabel('phi')
matplotlib.pyplot.ylabel('V')

matplotlib.pyplot.show()
```

Третий вариант:
$$\phi'_{n+1} = \phi'_n - \omega^2\cdot\sin{\phi_{n}}\cdot dt$$
$$\phi_{n+1} = \phi_n + \phi'_{n + 1}\cdot dt$$

```python
for i in range(1, n):
    V[i] = V[i - 1] - w*numpy.sin(phi[i - 1])*dt
    phi[i] = phi[i - 1] + V[i]*dt
```

```python
matplotlib.pyplot.figure(figsize=(15, 10))
matplotlib.pyplot.plot(phi, V, 'k')
matplotlib.pyplot.grid(True)
matplotlib.pyplot.axis('tight')
matplotlib.pyplot.xlabel('phi')
matplotlib.pyplot.ylabel('V')

matplotlib.pyplot.show()
```
